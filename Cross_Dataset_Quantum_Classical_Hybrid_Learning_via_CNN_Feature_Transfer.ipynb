```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, 1)
        self.conv2 = nn.Conv2d(32, 64, 3, 1)
        self.fc1 = nn.Linear(64 * 14 * 14, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
device = torch.device("cpu")
cnn_model = SimpleCNN()
cnn_model = cnn_model.to(device)
cnn_model.eval()
```

```python
from torch.utils.data import Dataset, DataLoader
from PIL import Image

class HFToTorchDataset(Dataset):
    def __init__(self, hf_dataset, transform=None):
        self.dataset = hf_dataset
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img = self.dataset[idx]['img']  # PIL image
        label = self.dataset[idx]['label']
        if self.transform:
            img = self.transform(img)
        return img, label

cifar_train_dataset = HFToTorchDataset(cifar['train'], transform=transform)
cifar_test_dataset  = HFToTorchDataset(cifar['test'],  transform=transform)

train_loader_cifar = DataLoader(cifar_train_dataset, batch_size=32, shuffle=True)
test_loader_cifar  = DataLoader(cifar_test_dataset,  batch_size=32, shuffle=False)
```

```python
#Defining & Training CNN feature Extractor
class CNNFeatureExtractor(nn.Module):
    def __init__(self):
        super(CNNFeatureExtractor, self).__init__()
        self.features = models.resnet18(pretrained=True)  # transfer learning
        self.features.fc = nn.Identity()  # remove final classification layer

    def forward(self, x):
        return self.features(x)

device = torch.device("cpu")

cnn_model = CNNFeatureExtractor().to(device)

# For speed: freeze layers
for param in cnn_model.parameters():
    param.requires_grad = False
```

```python
#Feature Extraction
def extract_features(loader):
    all_features = []
    all_labels = []
    with torch.no_grad():
        for imgs, labels in loader:
            imgs = imgs.to(device)
            feats = cnn_model(imgs)
            all_features.append(feats.cpu().numpy())
            all_labels.append(labels.numpy())
    return np.vstack(all_features), np.hstack(all_labels)

cifar_features, cifar_labels = extract_features(test_loader_cifar)
stl_features, stl_labels     = extract_features(stl_loader)
```

```python
#Dimensionality Reduction
scaler = StandardScaler()
cifar_features_scaled = scaler.fit_transform(cifar_features)
stl_features_scaled   = scaler.transform(stl_features)

pca = PCA(n_components=4)
cifar_features_pca = pca.fit_transform(cifar_features_scaled)
stl_features_pca   = pca.transform(stl_features_scaled)
```

```python
# Hybrid Quantum-Classical Neural Network Setup
from qiskit_machine_learning.neural_networks import EstimatorQNN
from qiskit.primitives import Estimator
from qiskit.circuit.library import TwoLocal, ZZFeatureMap
from qiskit_machine_learning.connectors import TorchConnector

# 1. Define quantum circuits
num_qubits = 4
feature_map = ZZFeatureMap(num_qubits)  # For data encoding
var_circuit = TwoLocal(num_qubits, 'ry', 'cz', reps=1)  # Trainable part
full_circuit = feature_map.compose(var_circuit)  # Combined circuit

# 2. Create QNN
qnn = EstimatorQNN(
    circuit=full_circuit,
    estimator=Estimator(),
    input_params=feature_map.parameters,  # Input data parameters
    weight_params=var_circuit.parameters  # Trainable weights
)

# 3. Connect to PyTorch
qnn_torch = TorchConnector(qnn)
print("Hybrid QNN ready for training! Circuit parameters:", len(var_circuit.parameters))
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrowPatch

fig, ax = plt.subplots(figsize=(8, 3))
ax.axis('off')

# Define block positions and labels
blocks = [
    ("Feature Map\n(ZZFeatureMap)", 0, 0.1),
    ("Variational Circuit\n(TwoLocal Ansatz)", 3, 0.1),
    ("Measurement\n(Expectations)", 6, 0.1)
]

# Add blocks to the diagram
for label, x, y in blocks:
    rect = Rectangle((x, y), 2, 1, fill=True, color="#D3D3D3", ec="black")
    ax.add_patch(rect)
    ax.text(x + 1, y + 0.5, label, ha='center', va='center')

# Draw arrows between blocks
for i in range(len(blocks) - 1):
    start = (blocks[i][1] + 2, 0.6)
    end = (blocks[i+1][1], 0.6)
    arrow = FancyArrowPatch(start, end, connectionstyle="arc3", arrowstyle='->')
    ax.add_patch(arrow)

ax.set_xlim(-0.5, 9)
ax.set_ylim(0, 1.5)
plt.title("Simplified Hybrid Quantum‑Classical NN Flow", fontsize=14)
plt.show()

```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(20, 4),  # Classical preprocessing
    qnn_torch,              # Quantum layer
    torch.nn.Linear(4, 10)  # Classical postprocessing
)
```

```python
#Training Quantum Classifier    Training it on CIFAR-10 features
class QuantumClassifier(nn.Module):
    def __init__(self, qnn_layer):
        super().__init__()
        self.qnn_layer = qnn_layer
        self.fc = nn.Linear(1, 10)  # map QNN output to 10 classes

    def forward(self, x):
        x = self.qnn_layer(x)
        return self.fc(x)

model = QuantumClassifier(qnn_torch)
optimizer = optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.CrossEntropyLoss()

# Training loop (small subset for demo)
X_train = torch.tensor(cifar_features_pca[:500], dtype=torch.float32)
y_train = torch.tensor(cifar_labels[:500], dtype=torch.long)

for epoch in range(5):
    optimizer.zero_grad()
    output = model(X_train)
    loss = loss_fn(output, y_train)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
#Evaluate on STL-10
X_test = torch.tensor(stl_features_pca[:200], dtype=torch.float32)
y_test = torch.tensor(stl_labels[:200], dtype=torch.long)

with torch.no_grad():
    preds = model(X_test)
    acc = (preds.argmax(dim=1) == y_test).float().mean()
print(f"Test Accuracy on STL-10: {acc*100:.2f}%")
```
